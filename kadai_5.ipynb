```python
import matplotlib.pyplot as plt
import numpy as np

y0 = [1, 2, -5, 2]

x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

# 平均が15, 標準偏差5の標準正規分布から2000個の乱数の生成する
x1 = np.random .normal(15, 5, 2000)

np.random.seed(3)
x2 = 4 + np.random.normal(0, 2, 60)
y2 = 4 + np.random.normal(0, 2, len(x2))


# プロット
fig, axes = plt.subplots(2, 3, tight_layout=True)

axes[0,0].plot(y0)
axes[0,1].plot(x, y, linewidth=2.0)

axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

axes[0,2].hist(x1)

axes[1, 0].scatter(x2, y2)

plt.show()

```
